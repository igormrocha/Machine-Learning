# Categorical Variables

There are two types of categorical variables

Nominal: When there is no hierarchy between the categories. 
Examples: male/female, green/red/blue, high/medium/low

Ordinal: When there is a hierarchy between the categories.
Examples: graduate/masters/phd , high/medium/low

When you have nominal variables in your dataset you should create dummy variables columns and drop one of them (the last columns is not necessary). 

![title](one_hot_encoding.png)


This technique for handling nominal variables is called one hot encoding.

When the variables are ordinal, then label encoding should be used.

# K Fold Cross Validation

It is a way of dividing your dataset for training and testing. 

When the traditional method is used, the training and test datapoints are separated randomly and this may cause a problem. For example, suppose you have the digits dataset and when you split it the training dataset doesn't have 0s, the 0s are in the test dataset. The model is therefore not trained to identify 0s but is tested against 0s, so the performance won't be good in this case. But since the dataset is splitted randomly, every time that you execute the code the score will change.

When using K Fold Cross Validation the data is divided into K folds, then the model uses one of the folds for test and the others for training, then the test fold is changed and this is repeated until the last fold is used for test. The score of the model will be the average of the scores of each training/test dataset. This way we can solve the problem of changing the score after each execution and we make sure the all the slices of the dataset were used for training as well as test.  

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
digits = load_digits()
dir(digits)
```

```python
lr = LogisticRegression()
svm = SVC()
rf = RandomForestClassifier()
```

__Basic Example__

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=3)
kf
```

```python
for train_index, test_index in kf.split([1,8,4,5,9,6,8,9,1]):
    print(train_index, test_index)
```

__Doing it in the hard way__

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=4)

lr_scores = []
svm_scores = []
rf_scores = []

for train_index, test_index in folds.split(digits.data, digits.target):
    X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index], \
                                       digits.target[train_index], digits.target[test_index]
    lr_scores.append(get_score(lr,X_train, X_test, y_train, y_test))
    svm_scores.append(get_score(svm,X_train, X_test, y_train, y_test))
    rf_scores.append(get_score(rf,X_train, X_test, y_train, y_test))
    
lr_avg = sum(lr_scores)/len(lr_scores)
svm_avg = sum(svm_scores)/len(svm_scores)
rf_avg = sum(rf_scores)/len(rf_scores)

print(lr_avg, '\n', svm_avg, '\n', rf_avg)
```

__Easy way (cross_val_score function)__

```python
from sklearn.model_selection import cross_val_score 

cross_val_score(lr, digits.data, digits.target)

cross_val_score(svm, digits.data, digits.target)

cross_val_score(rf, digits.data, digits.target)
```

# Hyperparameter Tuning (Using GridSearchCV) 

In the example above, we tried three diferent classifiers using K Fold Cross validation to see which one gives us the best score. However,the parameters of the models weren't tuned to the ideal values and consequently the scores returned may not be the best scores they can achieve.

To adjust the parameters as well as to compare the different models, we are going to use __GridSearchCV__. It basically runs a for loop changing the specified parameters and returns the scores of the model in each iteration.  

__Example for only one Classifiers__

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

clf = GridSearchCV (
    estimator=SVC(),
    param_grid={        
        'C':range(1,11),
        'kernel':['rbf','linear','poly'],
        'gamma':['auto','scale']        
    },
    #n_iter = 10
)

clf.fit(digits.data,digits.target);
df = pd.DataFrame(clf.cv_results_)
df = df[['params', 'mean_test_score']]
df.sort_values(by = 'mean_test_score', ascending = False)
```

__When you want to do grid search for more than one classifer...__

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([('classifier',SVC())])
search_space=[{ 'classifier': [SVC()],
               'classifier__C': range(1,11),
               'classifier__kernel':['linear','poly','rbf'],
               'classifier__gamma':['auto','scale']
             },
             { 'classifier':[RandomForestClassifier()],
               'classifier__n_estimators':np.arange(1,102,10),
               'classifier__criterion':['gini','entropy']
             }]


clf = GridSearchCV(pipe,param_grid=search_space)

clf.fit(digits.data,digits.target);
df = pd.DataFrame(clf.cv_results_)
df = df[['params', 'mean_test_score']]
df.sort_values(by = 'mean_test_score', ascending = False)
df
best_model = clf.best_estimator_.get_params()['classifier']
```
